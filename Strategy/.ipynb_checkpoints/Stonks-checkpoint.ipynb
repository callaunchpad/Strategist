```python
import csv
import pandas as pd
import numpy as np

tickers = ["AAPL", "TSLA"]
stocks = {}

for ticker in tickers:
    stocks[ticker] = pd.read_csv("./archive/stocks/{}.csv".format(ticker))
```

```python
for ticker in tickers:
    d2 = stocks[ticker][['Close']].pct_change().pct_change()
    stocks[ticker]['d2'] = d2
    stocks[ticker] = stocks[ticker].replace([np.inf, -np.inf], np.nan).fillna(0).replace(np.nan, 0)
```
